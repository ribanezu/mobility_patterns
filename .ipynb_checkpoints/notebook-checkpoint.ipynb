```python
import matplotlib.pyplot as plt
import numpy as np

# Define the bus stops (10 stops)
stops = [f'Stop {i}' for i in range(1, 11)]
n_stops = len(stops)

# Data designed to ensure that the cumulative passenger count is zero at the final stop
passengers_board = np.array([20, 5, 3, 0, 0, 2, 0, 0, 0, 0])
passengers_alight = np.array([0, 0, 10, 5, 3, 5, 2, 1, 2, 2])  # Cumulative: 20, 25, 18, 13, 10, 7, 5, 4, 2, 0

# Define the color palette
color_boarding = '#F6BD60'   # Yellowish for boardings
color_alighting = '#F28482'  # Reddish for alightings
color_cumulative = '#84A59D' # Teal-like for cumulative line
color_grid = '#F5CAC3'       # Pinkish for grid lines
bg_color = '#F7EDE2'         # Light background for figure and axes

# Create a figure with two subplots and set the overall background color
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
fig.patch.set_facecolor(bg_color)

# --------------------- Graph 1: Boardings and Alightings per Stop ---------------------
x_pos = np.arange(n_stops)
bar_width = 0.5

# Setting background color for the first axis
ax[0].set_facecolor(bg_color)

# Plot boardings (in color_boarding) and alightings (in color_alighting, negative values)
ax[0].bar(x_pos, passengers_board, bar_width, color=color_boarding, label='Boarding')
ax[0].bar(x_pos, -passengers_alight, bar_width, color=color_alighting, label='Alighting')

# Customize the axes, labels and title in English
ax[0].set_xticks(x_pos)
ax[0].set_xticklabels(stops, fontsize=12)
ax[0].set_ylabel('Number of Passengers', fontsize=12)
ax[0].set_title('Boardings and Alightings per Stop', fontsize=14, fontweight='bold')
ax[0].legend(fontsize=12)
ax[0].grid(axis='y', linestyle='--', alpha=0.7, color=color_grid)

# --------------------- Graph 2: Cumulative Passenger Count ---------------------
# Calculate cumulative passengers on board (boardings minus alightings)
net_passengers = np.cumsum(passengers_board - passengers_alight)
# Add the initial state (0 passengers before starting)
net_passengers_line = np.concatenate(([0], net_passengers))

# Create an x-axis for the cumulative data, labelling the initial point as 'Start'
x_points = np.arange(len(net_passengers_line))
labels = ['Start'] + stops

# Set background color for the second axis
ax[1].set_facecolor(bg_color)

# Plot with a step function for clarity, with markers for each point, using the designated cumulative color
ax[1].step(x_points, net_passengers_line, where='post', color=color_cumulative, label='Cumulative Passengers')
ax[1].scatter(x_points, net_passengers_line, color=color_cumulative, s=50)

# Customize x-axis labels and other elements
ax[1].set_xticks(x_points)
ax[1].set_xticklabels(labels, fontsize=12)
ax[1].set_ylabel('Cumulative Passenger Count', fontsize=12)
ax[1].set_title('Cumulative Passenger Count on Bus', fontsize=14, fontweight='bold')
ax[1].grid(axis='both', linestyle='--', alpha=0.7, color=color_grid)
ax[1].legend(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ---------------------------- Datos y Paleta ----------------------------
# Define the bus stops (10 stops)
stops = [f'Stop {i}' for i in range(1, 11)]
n_stops = len(stops)

# Data for boardings and alightings, ensuring that cumulative passengers == 0 at the final stop.
# Order of operations: first, alightings are processed; then, boardings.
# (Here we assume that boardings occur after alighting at each stop.)
passengers_board = np.array([20, 5, 3, 0, 0, 2, 0, 0, 0, 0])
passengers_alight = np.array([0, 0, 10, 5, 3, 5, 2, 1, 2, 2])
# Cumulative check: 
# Stop1: 20, Stop2: 20+5=25, Stop3: 25+3-10=18, Stop4: 18-5=13, Stop5: 13-3=10,
# Stop6: 10+2-5=7, Stop7: 7-2=5, Stop8: 5-1=4, Stop9: 4-2=2, Stop10: 2-2=0

# Paleta de colores
color_boarding    = '#F6BD60'   # para boardings
color_alighting   = '#F28482'   # para alightings
color_cumulative  = '#84A59D'   # para la línea y puntos de acumulado
color_grid        = '#F5CAC3'   # para las líneas de grid
bg_color          = '#F7EDE2'   # fondo de la figura y ejes

# ---------------------------- Lógica para Asignar Viajes ----------------------------
# La idea es asignar, en orden, los pasajeros que deben bajarse en cada parada.
# Se mantiene una "cola" de pasajeros activos, que es una lista de diccionarios con:
#   'origin': la parada de ingreso,
#   'passengers': número de pasajeros aún a bajarse.
# Al llegar a cada parada, se "desembarca" la cantidad indicada, sacándolos (first in, first out).

active = []  # lista de pasajeros a bordo, cada elemento: {'origin': str, 'passengers': int}
journeys = []  # lista de registros con origin, destination y número de pasajeros

for i in range(n_stops):
    stop = stops[i]
    # Primero se procesan los desembarques (alightings) en la parada i.
    alight = passengers_alight[i]
    while alight > 0 and active:
        # Toma el primer grupo (first in, first out)
        current = active[0]
        origin = current['origin']
        available = current['passengers']
        # Asignamos tantos pasajeros como sea posible para bajarse en esta parada.
        assign = min(available, alight)
        journeys.append({'origin': origin, 'destination': stop, 'passengers': assign})
        # Reducimos la cantidad pendiente y el total alight
        current['passengers'] -= assign
        alight -= assign
        if current['passengers'] == 0:
            active.pop(0)
    # Luego se procesan los boardings en la misma parada
    board = passengers_board[i]
    if board > 0:
        active.append({'origin': stop, 'passengers': board})

# Convertir la lista de journeys a un DataFrame
df_journeys = pd.DataFrame(journeys)

# Mostrar el DataFrame resultante
print("DataFrame of journeys:")
print(df_journeys)

# ---------------------------- Gráficos ----------------------------
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
fig.patch.set_facecolor(bg_color)

# Graph 1: Boardings and Alightings per Stop
x_pos = np.arange(n_stops)
bar_width = 0.5

ax[0].set_facecolor(bg_color)
ax[0].bar(x_pos, passengers_board, bar_width, color=color_boarding, label='Boarding')
ax[0].bar(x_pos, -passengers_alight, bar_width, color=color_alighting, label='Alighting')
ax[0].set_xticks(x_pos)
ax[0].set_xticklabels(stops, fontsize=12)
ax[0].set_ylabel('Number of Passengers', fontsize=12)
ax[0].set_title('Boardings and Alightings per Stop', fontsize=14, fontweight='bold')
#ax[0].legend(fontsize=12)
ax[0].grid(axis='y', linestyle='--', alpha=0.7, color=color_grid)

# Graph 2: Cumulative Passenger Count on Bus
# Calculate cumulative passengers on board after each stop.
net_passengers = np.cumsum(passengers_board - passengers_alight)
net_passengers_line = np.concatenate(([0], net_passengers))  # include initial 0

x_points = np.arange(len(net_passengers_line))
labels = ['Start'] + stops

ax[1].set_facecolor(bg_color)
ax[1].step(x_points, net_passengers_line, where='post', color=color_cumulative, label='Cumulative Passengers')
ax[1].scatter(x_points, net_passengers_line, color=color_cumulative, s=50)
ax[1].set_xticks(x_points)
ax[1].set_xticklabels(labels, fontsize=12)
ax[1].set_ylabel('Cumulative Passenger Count', fontsize=12)
ax[1].set_title('Cumulative Passenger Count on Bus', fontsize=14, fontweight='bold')
ax[1].grid(axis='both', linestyle='--', alpha=0.7, color=color_grid)
#ax[1].legend(fontsize=12)

plt.tight_layout()
plt.show()
```
